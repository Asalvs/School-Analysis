```python
import pandas as pd
```

```python
#address the files
school_data="Resources/schools_complete.csv"
student_data="Resources/students_complete.csv"
#Read Files
school_data_df=pd.read_csv(school_data)
student_data_df=pd.read_csv(student_data)
```

```python
#combine into the one dataset
school_data_complete_df=pd.merge(student_data_df , school_data_df , on="school_name", how="left")
school_data_complete_df.head(5)
```

```python
#LGA Summary

#total students
total_students=school_data_complete_df["student_name"].count()

#Total schools
total_schools=school_data_complete_df["school_name"].nunique()

#Total budget
total_budget=school_data_df["budget"].sum()

#Average maths score
average_maths_score=school_data_complete_df["maths_score"].sum()/total_students

#Average reading score
average_reading_score=school_data_complete_df["reading_score"].sum()/total_students

#Percentage passing math score
passing_maths_score_df=school_data_complete_df.loc[school_data_complete_df["maths_score"] >= 50]
percentage_passing_maths_score=len(passing_maths_score_df)/total_students * 100

#Percentage passing reading score
passing_reading_score_df=school_data_complete_df.loc[school_data_complete_df["reading_score"] >= 50]
percentage_passing_reading_score=len(passing_reading_score_df)/total_students * 100

#Percentage overall passing score
overall_passing_score_df=school_data_complete_df.loc[(school_data_complete_df["maths_score"] >= 50)\
                                                      & (school_data_complete_df["reading_score"] >= 50)]
percentage_overall_passing=len(overall_passing_score_df)/total_students * 100
```

```python
LGA_summary_df=pd.DataFrame({"Total Schools":[total_schools], 
                             "Total Students": total_students ,
                             "Total Budget": total_budget , 
                             "Average Maths Score": average_maths_score,
                             "Average Reading Score": average_reading_score , 
                             "% Passing Maths": percentage_passing_maths_score,
                             "% Passing Reading": percentage_passing_reading_score, 
                             "% Overall Passing": percentage_overall_passing })

#Format total budget to have a cleaner look                            
LGA_summary_df["Total Budget"]=LGA_summary_df["Total Budget"].map("${:,.2f}".format)
LGA_summary_df[["Total Schools",
      "Total Students",
      "Total Budget",
      "Average Maths Score",
      "Average Reading Score",
      "% Passing Maths",
      "% Passing Reading",
      "% Overall Passing"]]
```

```python
#School Summary
```

```python
#group our data frame by school name
school_group_df=school_data_complete_df.groupby(["school_name"])

#school type
school_type=school_group_df["type"].first()

#total students
total_student=school_group_df["Student ID"].count()

#total budget
total_budget=school_group_df["budget"].first()

#budget per student
budget_per_student= total_budget/total_student

#Average maths score
Avg_maths_score=school_group_df["maths_score"].mean()

#Average reading score
Avg_reading_score=school_group_df["reading_score"].mean()

#percentage passing math
passing_maths=school_data_complete_df[school_data_complete_df["maths_score"] >= 50].groupby(["school_name"])\
            ["maths_score"].count()
Percent_passing_maths = (passing_maths / total_student)*100

#Percentage reading pass
passing_reading=school_data_complete_df[school_data_complete_df["reading_score"] >= 50].groupby(["school_name"])\
            ["reading_score"].count()
Percent_passing_reading = (passing_reading / total_student)*100

#Percentage overal passing
overal_passing= school_data_complete_df[(school_data_complete_df["maths_score"] >= 50)\
                                                      & (school_data_complete_df["reading_score"] >= 50)]\
                                                        .groupby(["school_name"])["Student ID"].count()

percentage_overal_passing=(overal_passing/total_student)*100
```

```python
# create a data frame to hold all those info
school_summary_df = pd.DataFrame({"School Type": school_type,
      "Total Students": total_student,
      "Total School Budget": total_budget,
      "Per Student Budget": budget_per_student,
      "Average Math Score": Avg_maths_score,
      "Average Reading Score": Avg_reading_score,
      "% Passing Math": Percent_passing_maths,
      "% Passing Reading": Percent_passing_reading,
      "% Overall Passing": percentage_overal_passing})

#Format "total school budget" and "per student budget" to have a cleaner look  

school_summary_df["Total School Budget"]=school_summary_df["Total School Budget"].map("${:,.2f}".format)
school_summary_df["Per Student Budget"]=school_summary_df["Per Student Budget"].map("${:,.2f}".format)
school_summary_df.index.name= None

school_summary_df[["School Type",
      "Total Students",
      "Total School Budget",
      "Per Student Budget",
      "Average Math Score",
      "Average Reading Score",
      "% Passing Math",
      "% Passing Reading",
      "% Overall Passing"]]
```

```python
#Highest-Performing Schools (by % Overall Passing)
top_schools_df=school_summary_df.sort_values("% Overall Passing" , ascending=False)
top_schools_df.index.name= None
top_schools_df.head(5)
```
